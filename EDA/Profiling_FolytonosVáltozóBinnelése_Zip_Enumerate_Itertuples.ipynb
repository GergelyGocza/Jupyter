### Folytonos változó binnelése

```python
performance = pd.cut(df['Horsepower'], bins=[0, 80, 150, 250], labels=['Gyenge', 'Közepes', 'Erős'])
performance
```

#### Binnelt értékekből új változó létrehozása (új oszlop)

```python
performance = pd.Series(performance)
performance

df = pd.concat([df, performance.rename('Performance')], axis=1)
df.head()
```

### Iterálás - zip

```python
for le, sz in zip(df['Horsepower'], df['Origin']):
    if le > 200:
        print(sz)
```

### Iterálás - enumerate

```python
for i, x in enumerate(df['Horsepower']):
    if x > 200:
        df.at[i, 'Valami'] = df.at[i, 'Miles_per_Gallon']+1
        
df[df['Horsepower'] > 190]
```

### Iterálás - itertuples

```python
for row in df.itertuples(): 
    if(getattr(row, 'Horsepower') > 200):
        print(getattr(row, 'Index'), getattr(row, "Horsepower"), getattr(row, "Performance"))
```
