### Könyvtárak importálása

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Növekvő minta elemszám mellett a minták átlagainak követése

```python
x = np.arange(1, 2000, 1)
dobasok = []

for i in x:
    dobasok.append(np.random.randint(1, 7, i))
        
atlagok = []

for i in dobasok:
    atlagok.append(np.array(i).mean())
```

### Ábrázolás

```python
plt.figure(figsize=(12, 7))
plt.plot(x, atlagok)

plt.title('Nagy számok törvénye', size=15)
plt.xlabel('Minta elemszám')
plt.ylabel('Minta átlag')
```
