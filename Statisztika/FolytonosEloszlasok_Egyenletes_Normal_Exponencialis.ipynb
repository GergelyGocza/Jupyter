### Könyvtárak importálása

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

###  
###  
### Egyenletes eloszlás

```python
from scipy.stats import uniform

kezdő = 2
terjedelem = 5

#X tengely
dist_egyenletes_x = uniform.rvs(loc=kezdő, scale=terjedelem, size=100) #loc=átlag, scale=szórás
dist_egyenletes_x.sort()

#Y tengely
dist_egyenletes_y = uniform.pdf(dist_egyenletes_x)
```

#### Ábrázolás

```python
plt.figure(figsize=(12, 7))
plt.plot(dist_egyenletes_x, dist_egyenletes_y, 'b-', lw=5, alpha=0.6)

plt.title('Egyenletes eloszlás', size=15)
```

#### Valószínűség

```python
print(f'p(X < 4): {uniform(kezdő, terjedelem).cdf(4)}')
print(f'p(3 < X < 6): {uniform(kezdő, terjedelem).cdf(6) - uniform(kezdő, terjedelem).cdf(3)}')
print(f'p(X > 1): {1 - uniform(kezdő, terjedelem).cdf(6)}')
```

### Normális eloszlás

```python
from scipy.stats import norm

átlag = 0
szórás = 1

#X tengely
dist_norm_x = norm.rvs(loc=átlag, scale=szórás, size=100) #loc=átlag, scale=szórás
dist_norm_x.sort()

#Y tengely
dist_norm_y = norm.pdf(dist_norm_x)
```

#### Ábrázolás

```python
plt.figure(figsize=(12, 7))
plt.plot(dist_norm_x, dist_norm_y, 'b-', lw=5, alpha=0.6)

plt.title('Normális eloszlás', size=15)
```

#### Valószínűség

```python
print(f'p(X < -1): {norm(átlag, szórás).cdf(-1)}')
print(f'p(-1 < X < 1): {norm(átlag, szórás).cdf(1) - norm(átlag, szórás).cdf(-1)}')
print(f'p(X > 1): {1 - norm(átlag, szórás).cdf(-1)}')
```

### Exponenciális eloszlás

```python
from scipy.stats import expon

átlag = 0
szórás = 1

#X tengely
dist_exp_x = expon.rvs(loc=átlag, scale=szórás, size=100)
dist_exp_x.sort()

#Y tengely
dist_exp_y = expon.pdf(dist_exp_x)
```

```python
plt.figure(figsize=(12, 7))
plt.plot(dist_exp_x, dist_exp_y, 'b-', lw=5, alpha=0.6)

plt.title('Exponenciális eloszlás', size=15)
```

#### Valószínűség

```python
print(f'p(X < 1): {expon(átlag, szórás).cdf(1)}')
print(f'p(1 < X < 2): {expon(átlag, szórás).cdf(2) - expon(átlag, szórás).cdf(1)}')
print(f'p(X > 1): {1 - expon(átlag, szórás).cdf(1)}')
```
