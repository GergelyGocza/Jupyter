### Könyvtárak importálása

```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

###  
### Hipergeometriai-eloszlás

```python
from scipy.stats import hypergeom

N = 20  #összes elem
K = 9   #kijelöltek
n = 5   #kiválasztás
k = 2   #kiválasztottak közül kijelöltek
```

#### Ábrázolás

```python
x = np.arange(0, n+1)
hypergeom(N, K, n).pmf(x)


plt.figure(figsize=(12, 7))
plt.plot(x, hypergeom(N, K, n).pmf(x), 'bo')
plt.vlines(x, 0, hypergeom(N, K, n).pmf(x), colors='b', lw=5, alpha=0.5) #lw=linewidth, alpha=áttetszőség

plt.title('Hipergeometriai eloszlás', size=20)
plt.xlabel('Lehetséges kimenetek')
plt.ylabel('Kimenetek valószínűségei')
```

#### Várhatő érték (E(X)) és szórás (D(X))

```python
print(f'Várható érték: {hypergeom.stats(N, K, n)[0]}')
print(f'Szórás: {hypergeom.stats(N, K, n)[1]}')
```

#### Valószínűség

Egy esemény (N=)20 esetből (K=)9-szer következett be. Mekkora a valószínűsége, hogy (n=)5 esetből (k=)2-szer következik be?

```python
Egy esemény (N=)20 esetből (K=)9-szer következett be. Mekkora a valószínűsége, hogy (n=)5 esetből (k=)2-szer következik be
```

```python
print(f'p = {hypergeom.pmf(k, N, K, n)}')
```

### Binimiális-eloszlás  -- k korlátos

#### Együtthatók megadása

```python
from scipy.stats import binom

p = 0.12  #esemény bekövetkezés valósíznűsége
n = 10    #ennyi esetből...
k = 2     #...ennyiszer következik be az esemény
```

#### Ábrázolás

```python
x = np.arange(binom.ppf(0.001, n, p), binom.ppf(0.9999, n, p)) #hány lehetséges kimenetet mutasson

plt.figure(figsize=(12, 7))
plt.plot(x, binom.pmf(x, n, p), 'bo', ms=8)
plt.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)

plt.title('Binomiális eloszlás', size=20)
plt.xlabel('Lehetséges kimenetek')
plt.ylabel('Kimenetek valószínűségei')
```

#### Várhatő érték (E(X)) és szórás (D(X))

```python
print(f'Várható érték: {binom.stats(n, p)[0]}')
print(f'Szórás : {binom.stats(n, p)[1]}')
```

#### Valószínűség

Egy esemény bekövetkezésének valószínűsége (p=)0,12. Ekkor (n=)10 esetből mekkora a valószínűsége, hogy pont (k=)2-szer következik be?

```python
print(f'p = {binom.pmf(k, n, p)}')
```

### Poisson-eloszlás -- k nem korlátos

```python
from scipy.stats import poisson

p = 2/15  #esemény bekövetkezés valósíznűsége
n = 15    #vizsgált időegység
lam = n*p #lambda = a vizsgált időegység(n) alatt hány darab átlagosan = n*p
k = 3     #ennyiszer következik be az esemény

x = np.arange(poisson.ppf(0.01, lam), poisson.ppf(0.999, lam)) #hány lehetséges kimenetet mutasson

plt.figure(figsize=(12, 7))
plt.plot(x, poisson.pmf(x, lam), 'bo', ms=8, label='poisson pmf')
plt.vlines(x, 0, poisson.pmf(x, lam), colors='b', lw=5, alpha=0.5)

plt.title('Poisson eloszlás', size=20)
plt.xlabel('Lehetséges kimenetek')
plt.ylabel('Kimenetek valószínűségei')
```

#### Várható érték (E(X)) és szórás (D(X))

```python
print(f'Várható érték: {poisson.stats(n, p)[0]}')
print(f'Szórás: {poisson.stats(n, p)[1]}')
```

#### Valószínűség

Egy esemény bekövetkezésének valószínűsége (p=)2/15. Ekkor (n=)15 esetből mekkora a valószínűsége, hogy pont (k=)3-szor következik be?

```python
print(f'p = {poisson.pmf(k, lam)}')
```
