```python
sns.relplot(data=df, x='A', y='C', kind='scatter')
```

```python
import datashader as ds
import datashader.glyphs
import datashader.transfer_functions as tf

from holoviews.operation.datashader import datashade, dynspread

cvs = ds.Canvas(plot_width=1000, plot_height=500)

scatter_agg = cvs.points(df, 'C', 'A') 
scatter_img = tf.shade(scatter_agg, cmap=['lightblue', 'darkblue'], how='eq_hist') #how=linear #how=eq_hist 
scatter_img
#tf.set_background(scatter_img, 'black')

ds.transfer_functions.dynspread(scatter_img, threshold=0.6, max_px=1, shape='circle', how='over', name=None)
```
