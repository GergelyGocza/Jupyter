```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def rolldice():
    
    dice = np.random.randint(1,101)
    
    if dice <=51:     #Ha 1 és 51 közötti számot dobunk, akkor vesztünk
        return False
    elif dice <=100:  #Ha 52 és 100 közötti számot dobunk, akkor nyerünk
        return True
```

```python
def play(pénz, tét, játékok_száma):
    Play_num = []
    Funds = []
    Final_funds = []

    for i in range(játékok_száma):
        if rolldice() == True:     #Ha nyerünk, akkor a pénzünket növeljük a téttel
            pénz += tét
            Play_num.append(i)
            Funds.append(pénz)
        else:                     #Ha vesztünk, akkor a pénzünket csökkentjük a téttel
            pénz -= tét 
            Play_num.append(i)
            Funds.append(pénz)

    plt.plot(Play_num, Funds)
    Final_funds.append(Funds[-1])
    return(Final_funds)
```

```python
play(10000, 100, 5)
```

```python
plt.figure(figsize=(12, 7))

ending_fund = []

for i in range(100):
    ending_fund.append(play(10000, 100, 10000))

plt.ylabel('Pénz')
plt.xlabel('Fogadások száma')
plt.show()

print('A játék végén a pénzünk: ' + str(np.array(ending_fund).mean()))
```

```python
plt.figure(figsize=(12, 7))
print(play(10000, 100, 10))
plt.plot(10000, color='red')
```

### Kör területének kiszámítása

```python
n = 100000
x = np.random.rand(n, 2)
kör = x[np.sqrt((x[:, 0]-0.5)**2 + (x[:, 1]-0.5)**2) < 0.5]
becslés = len(kör)/len(x)
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(x[:, 0], x[:, 1], s=0.5)
plt.scatter(kör[:, 0], kör[:, 1], color='red', s=0.5)
print(becslés)
```
