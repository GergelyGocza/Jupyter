```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X1 = np.array([1, 2, 3, 4, 5, 6, 7])
Y = np.array([3, 5, 5, 6, 6, 8, 9])
```

```python
plt.figure(figsize=(12, 7))
plt.scatter(X1, Y)
```

```python
b0 = 0
b1 = 0
```

```python
Y_ = b0 + b1*X1
Y_
```

```python
plt.figure(figsize=(12, 7))
plt.scatter(X1, Y)
plt.plot(X1, Y_, color='red')
```

```python
hiba = []
for i in range(3):
    rnd = np.random.randint(0, len(X1))
    hiba.append((Y_[rnd] - Y[rnd])**2)
np.sum(hiba)
```

```python
hibakovetes = []
```

```python
alpha = 0.001

for i in range(1000):
    iksz = []
    ipszilon = []
    for i in range(3):
        rnd = np.random.randint(0, len(X1))
        iksz.append(X1[rnd])
        ipszilon.append(Y[rnd])
    
    b0 = b0 - np.sum(2*(b0 + b1*np.array(iksz) - np.array(ipszilon)))*alpha
    b1 = b1 - np.sum(2*(b0 + b1*np.array(iksz) - np.array(ipszilon))*np.array(iksz))*alpha
    hibakovetes.append(np.sum((b0 + b1*X1 - Y)**2))
    
Y_ = b0 + b1*X1
print('b0: ', b0, ' b1: ', b1)
print('hiba négyzetösszeg: ', np.sum((Y_ - Y)**2))
```

```python
plt.figure(figsize=(12, 7))
plt.scatter(np.linspace(0, len(hibakovetes), len(hibakovetes)), hibakovetes)
```
