```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X1 = np.array([1, 2, 3, 4, 5, 6, 7])
Y = np.array([3, 5, 5, 6, 6, 8, 9])
```

```python
plt.figure(figsize=(12, 7))
plt.scatter(X1, Y)
```

```python
b0 = 0
b1 = 0
```

```python
Y_ = b0 + b1*X1
Y_
```

```python
plt.figure(figsize=(12, 7))
plt.scatter(X1, Y)
plt.plot(X1, Y_, color='red')
```

```python
rnd = np.random.randint(0, len(X1))
hiba = (Y_[rnd] - Y[rnd])**2
hiba
```

```python
hibakovetes = []
```

```python
alpha = 0.01

for i in range(1000):
    rnd = np.random.randint(0, len(X1))
    hiba = (b0+b1*X1[rnd] - Y[rnd])**2
    
    b0 = b0 - (2*(b0+b1*X1[rnd] - Y[rnd]))*alpha
    b1 = b1 - (2*(b0+b1*X1[rnd] - Y[rnd])*X1[rnd])*alpha
    hibakovetes.append(np.sum((b0 + b1*X1 - Y)**2))
    
Y_ = b0 + b1*X1
print('b0: ', b0, ' b1: ', b1)
print('hiba négyzetösszeg: ', np.sum((Y_ - Y)**2))
```

```python
plt.figure(figsize=(12, 7))
plt.scatter(np.linspace(0, len(hibakovetes), len(hibakovetes)), hibakovetes)
```
