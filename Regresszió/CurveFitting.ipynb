### Könyvtárak importálása

```python
import pandas as pd
import numpy as np
import researchpy as rp
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

### Exponenciális regresszió

```python
from scipy.optimize import curve_fit 

X = np.linspace(0, 1, num = 40) 
y = 3.45 * np.exp(1.334 * X) + np.random.normal(size = 40) 
  
def exp(x, a, b): 
    return a*np.exp(b*x) 
  
param, param_cov = curve_fit(exp, X, y) 
```

#### Exponenciális egyenlet együtthatói

```python
print('funcion coefficients:') 
print(param) 
print('Covariance of coefficients:')
print(param_cov)
```

#### Ábrázolás

```python
ans = (param[0]*(np.exp(param[1]*X))) 

plt.figure(figsize=(12, 7))
plt.plot(X, y, 'o', color ='red', label ="data") 
plt.plot(X, ans, '--', color ='blue', linewidth=2, label ="prediction") 

plt.legend() 
plt.show() 
```

#### Prediction

```python
def predict_exp(x):
    return param[0]*(np.exp(param[1]*x))

predict_exp(0.6)
```

#### R^2 score

```python
from sklearn.metrics import r2_score

yexp_predicted = list(map(predict_exp, X))
r2_score(y, yexp_predicted)
```

### Logaritmikus regresszió

```python
X = np.linspace(1, 100, num = 40) 
y = 3.45 * np.log(1.334 * X) + np.random.normal(size = 40) 

def log(x, a, b): 
    return a*np.log(b*x) 
  
param, param_cov = curve_fit(log, X, y) 

print('funcion coefficients:') 
print(param) 
print('Covariance of coefficients:') 
print(param_cov) 
```

#### Ábrázolás

```python
ans = (param[0]*(np.log(param[1]*X))) 

plt.figure(figsize=(12, 7))
plt.plot(X, y, 'o', color ='red', label ="data") 
plt.plot(X, ans, '--', color ='blue', linewidth=2, label ="optimized data") 

plt.legend() 
plt.show() 
```

#### Prediction

```python
def predict_log(x):
    return param[0]*(np.log(param[1]*x))

predict_log(60)
```

#### R^2 score

```python
from sklearn.metrics import r2_score

ylog_predicted = list(map(predict_log, X))
r2_score(y, ylog_predicted)
```
