### Könyvtárak importálása

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import os

ldir = os.chdir(r'C:\Users\Gergely\Documents\Datasets')
```

### DataFrame

```python
df = pd.read_csv('Life Expectancy Data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['Alcohol', 'Hepatitis B', 'Total expenditure'], inplace=True)
```

```python
df.columns
```

### Style beállítás

```python
plt.style.use('ggplot')
```

### Egy ábrán több vonal (y azonos típusú)

```python
fig, ax = plt.subplots()

ax.plot(df[df['Country'] == 'Hungary'].Year, df[df['Country'] == 'Hungary']['Life expectancy '], label='Magyarország')
ax.plot(df[df['Country'] == 'Germany'].Year, df[df['Country'] == 'Germany']['Life expectancy '], label='Németország')

ax.set_xlabel('Év')
ax.set_ylabel('Várható élettartam')
ax.set_title('Várható élettartam alakulása évenként')
ax.legend()

fig.set_figheight(7)
fig.set_figwidth(14)
```

### Egy ábrán több vonal (y eltérő típusú)

```python
fig, ax = plt.subplots()

ax.plot(df[df['Country'] == 'Hungary'].Year, df[df['Country'] == 'Hungary']['Life expectancy '], color='b', label='Várható élettartam')
ax.set_xlabel('Év')
ax.set_ylabel('Várható élettartam', color='b')
ax.tick_params('y', colors='b')

ax = ax.twinx()

ax.plot(df[df['Country'] == 'Hungary'].Year, df[df['Country'] == 'Hungary']['Schooling'], color='r', label='Iskolázottság')
ax.set_ylabel('Iskolázottság', color='r')
ax.tick_params('y', colors='r')

ax.set_title('Várható élettartam és Iskolázottság')

fig.set_figheight(7)
fig.set_figwidth(14)
```

### Külön ábrán a vonalak (saját Y terjedelemmel)

```python
fig, ax = plt.subplots(2, 1)

ax[0].plot(df[df['Country'] == 'Hungary'].Year, df[df['Country'] == 'Hungary']['Life expectancy '], color='b')
ax[0].set_ylabel('Magyarország')
ax[0].set_title('Magyarország és Németország várható életkotának alakulása')

ax[1].plot(df[df['Country'] == 'Germany'].Year, df[df['Country'] == 'Germany']['Life expectancy '])
ax[1].set_xlabel('Év')
ax[1].set_ylabel('Németország')

fig.set_figheight(11)
fig.set_figwidth(11)
```

### Külön ábrán a vonalak (közös Y terjedelemmel)

```python
fig, ax = plt.subplots(2, 1, sharey=True)

ax[0].plot(df[df['Country'] == 'Hungary'].Year, df[df['Country'] == 'Hungary']['Life expectancy '], color='b')
ax[0].set_ylabel('Magyarország')
ax[0].set_title('Magyarország és Németország várható életkotának alakulása')

ax[1].plot(df[df['Country'] == 'Germany'].Year, df[df['Country'] == 'Germany']['Life expectancy '])
ax[1].set_xlabel('Év')
ax[1].set_ylabel('Németország')

fig.set_figheight(11)
fig.set_figwidth(11)
```
